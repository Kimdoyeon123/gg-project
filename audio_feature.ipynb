# 위 2가지 방식 중 하나로 담겨진 노래 id들로부터 spotify api 사용해서 audio feature 추출하고 json 파일에 저장하기

```python
import os
import json
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Spotify API 인증 설정
sp = spotipy.Spotify(auth_manager=SpotifyClientCredentials(
    client_id="e8df1edf03f9423c834d34a101d1cd3c",       # Spotify에서 발급받은 Client ID
    client_secret="99cfcdbf92524503aefbe45b07060d62"    # Spotify에서 발급받은 Client Secret
))

# 플레이리스트 ID를 통해 모든 트랙 ID 가져오는 함수 (페이징 처리 추가)
def get_playlist_track_ids(playlist_id):
    track_ids = []
    offset = 0  # 페이징 시작점
    limit = 100  # 한 번에 가져올 수 있는 최대 트랙 수
    
    while True:
        # Spotify API 요청 (페이징 처리)
        results = sp.playlist_tracks(playlist_id, offset=offset, limit=limit)
        tracks = results['items']
        
        # 트랙 ID 추출
        for item in tracks:
            track = item['track']
            if track:  # 트랙이 None이 아닌 경우에만 추가
                track_ids.append(track['id'])
        
        # 다음 페이지가 없으면 종료
        if results['next'] is None:
            break
        
        # 다음 페이지로 이동
        offset += limit

    return track_ids

# 필요한 오디오 피쳐를 가져오는 함수
def get_selected_audio_features(track_id):
    features = sp.audio_features(track_id)
    if features and features[0]:  # features[0]이 None이 아닌 경우
        audio_features = features[0]
        # 기존 + 추가 피쳐 포함
        selected_features = {
            "tempo": audio_features['tempo'],
            "valence": audio_features['valence'],
            "energy": audio_features['energy'],
            "danceability": audio_features['danceability'],
            "acousticness": audio_features['acousticness'],
            "speechiness": audio_features['speechiness'],
            "liveness": audio_features['liveness'],
            "instrumentalness": audio_features['instrumentalness'],  # 추가
            "loudness": audio_features['loudness'],                  # 추가
            "key": audio_features['key'],                            # 추가
            "mode": audio_features['mode'],                          # 추가
            "duration_ms": audio_features['duration_ms']             # 추가
        }
        return selected_features
    return None

# 트랙 ID를 통해 노래 제목 가져오기
def get_track_name(track_id):
    track_info = sp.track(track_id)
    if track_info:
        return track_info['name']
    return None

# 기존 JSON 파일을 읽어오는 함수
def load_audio_features_data(file_path):
    try:
        with open(file_path, 'r') as f:
            data = json.load(f)
    except FileNotFoundError:
        # 파일이 없는 경우 빈 딕셔너리 반환
        data = {}
    return data

# 업데이트된 데이터를 JSON 파일에 저장하는 함수
def save_audio_features_data(audio_features_data, file_path):
    try:
        with open(file_path, 'w') as f:
            json.dump(audio_features_data, f, indent=4)  # indent=4는 보기 좋게 들여쓰기
        print(f"데이터가 {file_path}에 성공적으로 저장되었습니다.")
    except Exception as e:
        print(f"파일을 저장하는 데 오류가 발생했습니다: {e}")

# 트랙 ID를 통해 오디오 피쳐를 추출하고 저장하는 함수
def process_tracks(track_ids, file_path):
    audio_features_data = load_audio_features_data(file_path)
    
    for track_id in track_ids:
        try:
            # 트랙 이름 가져오기
            track_name = get_track_name(track_id)
            if not track_name:
                print(f"트랙 이름을 가져올 수 없습니다: {track_id}")
                continue
            
            # 오디오 피쳐 가져오기
            selected_features = get_selected_audio_features(track_id)
            if selected_features:
                # JSON 데이터에 추가 (트랙 이름을 키로 사용)
                audio_features_data[track_name] = selected_features
                print(f"트랙 {track_name}의 오디오 피쳐가 업데이트되었습니다.")
                
                # 업데이트된 데이터 저장
                save_audio_features_data(audio_features_data, file_path)
            else:
                print(f"해당 트랙에 대한 오디오 피쳐를 가져올 수 없습니다: {track_name} ({track_id})")
        except Exception as e:
            print(f"오류 발생: {e} (트랙 ID: {track_id})")

# 플레이리스트 ID 입력 (실제 플레이리스트 ID로 대체하세요)
playlist_id = '24oihw9jD3T1oOXtjLwJug'

# 1. 플레이리스트에서 트랙 ID 가져오기
track_ids = get_playlist_track_ids(playlist_id)
print(f"플레이리스트에서 총 {len(track_ids)}개의 트랙 ID를 가져왔습니다.")

# 2. JSON 파일 경로 설정
file_path = os.path.abspath('data/sadness.json')
print(f"파일 경로 확인: {file_path}")

# 3. 오디오 피쳐 분석 및 JSON 파일 업데이트
process_tracks(track_ids, file_path)
```

```python
import json

# JSON 파일 경로 설정
json_files = {
    "anger": 'data/anger.json',
    "joy": 'data/joy.json',
    "neutral": 'data/neutral.json',
    "sadness": 'data/sadness.json'
}

# 모든 노래 제목과 값을 저장할 딕셔너리
all_data = {}

# 각 파일에서 노래 제목과 값을 추출하여 저장
for label, file_path in json_files.items():
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)
        # 각 JSON 파일의 데이터를 저장 (key-value 쌍)
        all_data[label] = data

# JSON 파일 간에 중복된 노래 찾기
duplicates = {}
labels = list(all_data.keys())

for i in range(len(labels)):
    for j in range(i + 1, len(labels)):
        label1 = labels[i]
        label2 = labels[j]

        # 중복된 키-값 쌍을 확인
        common_items = {
            key: all_data[label1][key]
            for key in all_data[label1]
            if key in all_data[label2] and all_data[label1][key] == all_data[label2][key]
        }

        if common_items:
            duplicates[(label1, label2)] = common_items

# 중복된 노래 출력
if duplicates:
    for (label1, label2), items in duplicates.items():
        print(f"{label1}와 {label2} 파일 간 중복된 노래들:")
        for title, value in items.items():
            print(f"  제목: {title}")
else:
    print("모든 JSON 파일 간 중복된 노래가 없습니다.")
```

```python
import numpy as np
import pandas as pd

# JSON 파일 경로
file_paths = {
    "anger": "data/anger.json",
    "joy": "data/joy.json",
    "neutral": "data/neutral.json",
    "sadness": "data/sadness.json"
}

# 피쳐별 데이터 제한값 계산
def calculate_feature_limits(data):
    feature_limits = {}
    for feature in data.columns:
        q1 = data[feature].quantile(0.25)  # 1사분위수
        q3 = data[feature].quantile(0.75)  # 3사분위수
        iqr = q3 - q1  # IQR
        lower_bound = q1 - 1.5 * iqr  # 하한값
        upper_bound = q3 + 1.5 * iqr  # 상한값
        feature_limits[feature] = (lower_bound, upper_bound)
    return feature_limits

# 각 감정별 제한값 계산 및 적용
feature_limits_by_emotion = {}
for emotion, file_path in file_paths.items():
    with open(file_path, "r") as f:
        data = json.load(f)

    # 데이터를 DataFrame으로 변환
    df = pd.DataFrame(data.values(), columns=[
        "tempo", "valence", "energy", "danceability", 
        "acousticness", "speechiness", "liveness", 
        "instrumentalness", "loudness", "key", "mode", "duration_ms"
    ])

    # 피쳐별 제한값 계산
    feature_limits = calculate_feature_limits(df)
    feature_limits_by_emotion[emotion] = feature_limits

    # 데이터 제한 적용
    for feature, (lower, upper) in feature_limits.items():
        df = df[(df[feature] >= lower) & (df[feature] <= upper)]

    # 제한된 데이터 저장
    filtered_data = {k: v for k, v in zip(data.keys(), df.values.tolist())}
    with open(file_path.replace(".json", "_filtered.json"), "w") as f:
        json.dump(filtered_data, f, indent=4)

    print(f"{emotion} 데이터 제한 완료: {len(filtered_data)}개 데이터 유지")
```
